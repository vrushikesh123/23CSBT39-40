```python
# Define the lists
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Combine the lists into a dataset
healthcare_data = []
for i in range(len(people)):
    person_data = {
        'Name': people[i],
        'Age': age[i],
        'Height (cm)': height[i],
        'Weight (kg)': weight[i]
    }
    healthcare_data.append(person_data)

# Print the dataset
for data in healthcare_data:
    print(data)

import matplotlib.pyplot as plt

# Dataset
people = ['sathwika', 'chinni', 'chinnu', 'assu']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]



# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(height, weight, color='blue', marker='o')

# Adding Titles and Labels
plt.title('Height vs Weight of Individuals')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# Display Plot
plt.grid(True)
plt.show()




# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(people, weight, color='green')

# Adding Titles and Labels
plt.title('Weight of Individuals')
plt.xlabel('Person')
plt.ylabel('Weight (kg)')

# Display Plot
plt.show()




# Histogram
plt.figure(figsize=(10, 6))
plt.hist(age, bins=4, color='orange', edgecolor='black')

# Adding Titles and Labels
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Display Plot
plt.show()
```
